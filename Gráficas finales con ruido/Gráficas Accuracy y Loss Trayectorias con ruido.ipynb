```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df200conv = pd.read_csv('History 100,200/historyconvtrayec200.csv')
df200dense = pd.read_csv('History 100,200/historytrayec200.csv')
df100conv = pd.read_csv('History 50,100//historyconvtrayec100.csv')
df100dense =  pd.read_csv('History 50,100/historytrayec100.csv')
df50conv = pd.read_csv('History 20,50/historyconvtrayec50.csv')
df50dense = pd.read_csv('History 20,50/historytrayec50.csv')
df50conv
```

```python
ax = df200conv['val_loss'].plot(kind='line', title='Loss y Val_Loss ruido', xlabel='Ã‰pocas', 
                                ylabel='PÃ©ridida en datos de validaciÃ³n', label='Conv 200')
df200dense['val_loss'].plot(kind='line', ax=ax, label='Densa 200')
df100dense['val_loss'].plot(kind='line', ax=ax, label='Densa 100')
df100conv['val_loss'].plot(kind='line', ax=ax, label='Conv 100')
df50conv['val_loss'].plot(kind='line', ax=ax, label='Conv 50')
df50dense['val_loss'].plot(kind='line', ax=ax, label='Densa 50')

# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```
