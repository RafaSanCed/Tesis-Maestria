```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df200 = pd.read_csv('historytrayec200.csv')
df100 = pd.read_csv('historytrayec100.csv')
df50 = pd.read_csv('historytrayec50.csv')
df200
```

```python
ax = df200['loss'].plot(kind='line', title='Loss y Val_Loss', xlabel='Eje X', ylabel='Eje Y', label='Loss 200')
df200['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 200')
df100['loss'].plot(kind='line', ax=ax, label='Loss 100')
df100['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 100')
df50['loss'].plot(kind='line', ax=ax, label='Loss 50')
df50['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 50')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```

```python
ax = df200['mae'].plot(kind='line', title='MAE y Val_MAE', xlabel='Eje X', ylabel='Eje Y', label='MAE 200')
df200['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 200')
df100['mae'].plot(kind='line', ax=ax, label='MAE 100')
df100['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 100')
df50['mae'].plot(kind='line', ax=ax, label='MAE 50')
df50['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 50')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```
