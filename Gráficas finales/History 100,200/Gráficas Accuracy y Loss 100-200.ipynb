```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfcampvec = pd.read_csv('historycampvec100.csv')
dfcosto = pd.read_csv('historycampvec100funcioncosto.csv')
dfconvcampvec = pd.read_csv('historyconvcampvec100.csv')
dfconvtray = pd.read_csv('historyconvtrayec200.csv')
dftray = pd.read_csv('historytrayec200.csv')
dfcosto
```

```python
ax = dfcosto['loss'].plot(kind='line', title='Loss', xlabel='Epochs', ylabel='Valor funciÃ³n de costo')
dfcampvec['loss'].plot(kind='line', ax=ax, label='Campo vectorial densa')
dfcosto['loss'].plot(kind='line', ax=ax, label='Campo vectorial funciÃ³n de costo')
dfconvcampvec['loss'].plot(kind='line', ax=ax, label='Convolucional campo vectorial')
dfconvtray['loss'].plot(kind='line', ax=ax, label='Convolucional trayectorias')
dftray['loss'].plot(kind='line', ax=ax, label='Trayectorias densa')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```

```python
ax = dfcosto['val_loss'].plot(kind='line', title='Val_Loss', xlabel='Epochs', ylabel='Valor funciÃ³n de costo datos validaciÃ³n')
dfcampvec['val_loss'].plot(kind='line', ax=ax, label='Campo vectorial densa')
dfcosto['val_loss'].plot(kind='line', ax=ax, label='Campo vectorial funciÃ³n de costo')
dfconvcampvec['val_loss'].plot(kind='line', ax=ax, label='Convolucional campo vectorial')
dfconvtray['val_loss'].plot(kind='line', ax=ax, label='Convolucional trayectorias')
dftray['val_loss'].plot(kind='line', ax=ax, label='Trayectorias densa')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```
