```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df100 = pd.read_csv('historycampvec100.csv')
df50 = pd.read_csv('historycampvec50.csv')
df20 = pd.read_csv('historycampvec20.csv')
df100
```

```python
ax = df100['loss'].plot(kind='line', title='Loss y Val_Loss', xlabel='Eje X', ylabel='Eje Y', label='Loss 100')
df100['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 100')
df50['loss'].plot(kind='line', ax=ax, label='Loss 50')
df50['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 50')
df20['loss'].plot(kind='line', ax=ax, label='Loss 20')
df20['val_loss'].plot(kind='line', ax=ax, label='Val_Loss 20')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```

```python
ax = df100['mae'].plot(kind='line', title='MAE y Val_MAE', xlabel='Eje X', ylabel='Eje Y', label='MAE 100')
df100['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 100')
df50['mae'].plot(kind='line', ax=ax, label='MAE 50')
df50['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 50')
df20['mae'].plot(kind='line', ax=ax, label='MAE 20')
df20['val_mae'].plot(kind='line', ax=ax, label='Val_MAE 20')
# Agrega leyenda y muestra la grÃ¡fica
ax.legend()
plt.show()
```
