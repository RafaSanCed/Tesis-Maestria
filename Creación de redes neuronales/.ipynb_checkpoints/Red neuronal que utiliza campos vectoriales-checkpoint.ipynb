# Mezclar y procesar por lotes los conjuntos de datos

```python
BATCH_SIZE = 1
SHUFFLE_BUFFER_SIZE = 1

train_dataset = train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
test_dataset = test_dataset.batch(BATCH_SIZE)
val_dataset = val_dataset.batch(BATCH_SIZE)
```

# Creamos función de costo

```python
def custom_loss(y_true,y_pred):
    campvec = []
    # constantes
    a1 = y_true[:,0]
    a1 = tf.cast(a1, tf.float64)
    
    a2 = y_true[:,1]
    a2 = tf.cast(a2, tf.float64)

    b1 = y_true[:,2]
    b1 = tf.cast(b1, tf.float64)

    b2 = y_true[:,3]
    b2 = tf.cast(b2, tf.float64)
    
    d1 = y_true[:,4]
    d1 = tf.cast(d1, tf.float64)
    
    d2 = y_true[:,5]
    d2 = tf.cast(d2, tf.float64)
    
    n = y_true[:,6]
    n = tf.cast(n, tf.float64)
    
    def f(t, y):

        # asignar a cada ODE a un elemento de vector
        X = y[0]
        #X = float(X)
        Y = y[1]
        #Y = float(Y)

        # definimos cada ODE
        dX_dt = a1/(1+tf.math.pow(Y,n))-d1*X+b1
        dY_dt = a2/(1+tf.math.pow(X,n))-d2*Y+b1

        return [dX_dt, dY_dt]

    x_range_neg = 0
    x_range_pos = 50
    y_range_neg = 0
    y_range_pos =50

    y1 = np.linspace(x_range_neg, x_range_pos, 100)
    y2 = np.linspace(y_range_neg, y_range_pos, 100)

    Y1, Y2 = np.meshgrid(y1, y2)
    
    t1 = 0

    u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

    NI, NJ = Y1.shape

    u1, v1 = f(t1, [Y1, Y2])
    u, v = f(t1, [Y1, Y2])
    #M = tf.math.sqrt(tf.square(u)+tf.square(u))
    M = (tf.experimental.numpy.hypot(u, v))
    u /= M
    v /= M

    campvecci = tf.experimental.numpy.append([u1], [v1], axis=0)

    campvec = campvecci

# ----------------------------------------------------------------------------
    campvec2 = []
    # constantes
    a12 = y_pred[:,0]
    a12 = tf.cast(a12, tf.float64)
    
    a22 = y_pred[:,1]
    a22 = tf.cast(a22, tf.float64)
    
    b12 = y_pred[:,2]
    b12 = tf.cast(b12, tf.float64)
    
    b22 = y_pred[:,3]
    b22 = tf.cast(b22, tf.float64)
    
    d12 = y_pred[:,4]
    d12 = tf.cast(d12, tf.float64)

    d22 = y_pred[:,5]
    d22 = tf.cast(d22, tf.float64)

    n2 = y_pred[:,6]
    n2 = tf.cast(n2, tf.float64)

    def f2(t2, y2):

        # asignar a cada ODE a un elemento de vector
        X2 = y2[0]
        Y2 = y2[1]

        # definimos cada ODE
        dX2_dt = a12/(1+tf.math.pow(Y2,n2))-d12*X+b12
        dY2_dt = a22/(1+tf.math.pow(X2,n2))-d22*Y+b12

        return [dX2_dt, dY2_dt]

    y12 = np.linspace(x_range_neg, x_range_pos, 100)
    y22 = np.linspace(y_range_neg, y_range_pos, 100)

    Y12, Y22 = np.meshgrid(y12, y22)
    
    t12 = 0

    u2, v2 = np.zeros(Y12.shape), np.zeros(Y22.shape)

    NI2, NJ2 = Y12.shape

    u12, v12 = f(t12, [Y12, Y22])
    u2, v2 = f(t12, [Y12, Y22])
    M2 = (tf.experimental.numpy.hypot(u2, v2))
    u2 /= M2
    v2 /= M2

    campvecci2 = tf.experimental.numpy.append([u12], [v12], axis=0)

    campvec2 = campvecci2

    
    return tf.reduce_mean(campvec-campvec2)
```

# Creación del modelo

```python
#model = Sequential()

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(2,100,100)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(7)
])
```

```python
model.summary()
```

# Entrenamos el modelo

```python
model.compile(optimizer='adam',
              loss= custom_loss,
              metrics=['accuracy'])
```

```python
def scheduler(epoch, lr):
  if epoch < 500:
    return lr
  else:
    return 1e-3 * 0.99 ** epoch
```

```python
val_epochs = 200
tf.compat.v1.global_variables_initializer()

early_stop = tf.keras.callbacks.EarlyStopping( monitor = 'val_loss', patience = 800,verbose = 1, 
                                              restore_best_weights = True)

reduce_lr = tf.keras.callbacks.LearningRateScheduler(scheduler)
history = model.fit(train_dataset, validation_data=val_dataset, epochs= val_epochs, callbacks=[reduce_lr, early_stop])
```

# Analizamos accuracy y loss

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(200)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
