```python
import random
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize
```

# Ajuste de red de regulación genética con una condición inicial

```python
# El sistema dinámico

a1 = random.randint(0, 10)
a2 = random.randint(0, 10)
b1 = random.randint(0, 10)
b2 = random.randint(0, 10)
d1 = random.randint(0, 10)
d2 = random.randint(0, 10)
n = random.randint(0, 5)
print(a1,a2,b1,b2,d1,d2,n)
def f(t,y):

    X = y[0]
    Y = y[1]

    dX_dt = a1/(1+Y**n)-d1*X+b1
    dY_dt = a2/(1+X**n)-d2*Y+b1

    return [dX_dt, dY_dt]

t_span = [0, 100]
times = np.linspace(t_span[0], t_span[1], 200)
#y0 = np.array([random.randint(0, 10), random.randint(0, 10)])
y0 = np.array([10,2])
sol = solve_ivp(f, t_span,y0,t_eval=times)
var1 = sol.y[0]
var2 = sol.y[1]

plt.figure(figsize=(5, 5))
plt.plot(times, var1, label = 'Concentración X')
plt.plot(times, var2, label = 'Concentración Y')
plt.legend()
plt.show()
```

```python
def F(T,a1,a2,b1,b2,d1,d2,n):

    t = T[:int(len(T)/2)]

    def f(t, y):

        X = y[0]
        Y = y[1]

        dX_dt = a1/(1+Y**n)-d1*X+b1
        dY_dt = a2/(1+X**n)-d2*Y+b1

        return [dX_dt, dY_dt]
    
    t_span = [0, 100]
    sol = solve_ivp(f,t_span,[10,2],t_eval=t)
    var1 = sol.y[0]
    var2 = sol.y[1]
    return np.concatenate([var1,var2],axis = 0)
```

```python
T = np.concatenate([times,times], axis = 0)
y = np.concatenate([var1, var2], axis = 0)

params, pcov = optimize.curve_fit(F, T, y, bounds=[0,10])
```
